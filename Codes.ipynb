### Primary Dataset

```python
primary = pd.read_csv(
    "data/allcauseunplanned30-dayhospitalreadmissionratecalifornia2011_2023.csv"
)
primary.head()
```

```python
primary = primary[(primary["Strata"] == "Overall") & (primary["County"] != "State")]
primary
```

```python
primary.columns
```

```python
primary["County"].unique()
```

```python
columns_to_drop = [
    "Strata",
    "Strata Name",
    "Total Admits (ICD-10)",
    "30-day Readmits (ICD-10) ",
    "30-day Readmission Rate (ICD-10)",
]
primary_pre2016 = primary[primary["Year"] < 2016]
primary_pre2016 = primary_pre2016.drop(columns=columns_to_drop)
new_column_names = {
    "Total Admits (ICD-9)": "Total_Admits",
    "30-day Readmits (ICD-9) ": "30_day_Readmits",
    "30-day Readmission Rate (ICD-9)": "30_day_Readmission_Rate",
}
primary_pre2016 = primary_pre2016.rename(columns=new_column_names)

primary_pre2016
```

```python
columns_to_drop_2016 = [
    "Strata",
    "Strata Name",
    "Total Admits (ICD-9)",
    "30-day Readmits (ICD-9) ",
    "30-day Readmission Rate (ICD-9)",
]
primary_after2016 = primary[primary["Year"] >= 2016]
primary_after2016 = primary_after2016.drop(columns=columns_to_drop_2016)
new_column_names_2016 = {
    "Total Admits (ICD-10)": "Total_Admits",
    "30-day Readmits (ICD-10) ": "30_day_Readmits",
    "30-day Readmission Rate (ICD-10)": "30_day_Readmission_Rate",
}
primary_after2016 = primary_after2016.rename(columns=new_column_names_2016)
primary_after2016
```

```python
cleaned_primary = pd.concat([primary_pre2016, primary_after2016], ignore_index=True)
cleaned_primary = cleaned_primary[cleaned_primary["Year"].isin([2019, 2020, 2021])]
cleaned_primary.drop(cleaned_primary[cleaned_primary['County'] == 'Unknown County'].index, inplace=True)
cleaned_primary
```

```python
cleaned_primary.to_csv("data/cleaned_data/primary.csv", index=False)
```

### Measure Dataset

```python
measures = pd.read_csv("data/measures-data-refresh2024.csv")
measures.head()
```

```python
drop_columns = ["county_name_plus_spa", "covered_california_region", "suppression_ind"]
measures = measures[measures["reporting_year"].isin([2019, 2020, 2021])]
measures = measures.drop(columns=drop_columns)
measures
```

```python
measures_new_column_names = {"reporting_year": "Year", "county_name": "County"}
measures = measures.rename(columns=measures_new_column_names)
measures
```

```python
grouped_measures = measures.groupby(["Year", "County", "measure_id"]).agg(
    Numerator=("measure_numerator", "sum"), Denominator=("measure_denominator", "sum")
)
grouped_measures.head(37)
```

```python
measures_signle_measure = measures[measures["measure_id"] == 3]
measures_signle_measure = measures_signle_measure.drop(columns=["measure_id"])
county_enrollment = measures_signle_measure.groupby(["Year", "County"]).agg(
    Enrollment=("measure_denominator", "sum")
)
```

```python
county_enrollment
```

```python
measures = measures[measures["measure_id"].isin([3, 5, 6, 9, 14, 15, 20, 32])]
measures
```

```python
merged_measures = pd.merge(
    measures, county_enrollment, on=["Year", "County"], how="left"
)
merged_measures
```

```python
measures_age = measures[measures["measure_id"].isin([3])]

measures_age = (
    measures_age.groupby(["Year", "County"])
    .agg(
        total_numerator=("measure_numerator", "sum"),
        total_denominator=("measure_denominator", "sum"),
    )
    .reset_index()
)

measures_age["Average_age"] = (
    measures_age["total_numerator"] / measures_age["total_denominator"]
)
measures_age = measures_age.drop(columns=["total_numerator", "total_denominator"])
measures_age["Average_age"] = measures_age["Average_age"].round(2)
measures_age
```

```python
def calculate_specific_percentage(merged_measures, measure_id, measure):

    measures_specific = merged_measures[
        merged_measures["measure_id"].isin([measure_id])
    ]

    # Group by Year and County, and calculate the sum of measure_numerator and first Enrollment
    measures_specific = (
        measures_specific.groupby(["Year", "County"])
        .agg(
            total_numerator=("measure_numerator", "sum"),
            Enrollment=("Enrollment", "first"),
        )
        .reset_index()
    )

    # Calculate the percentage of heart attack cases
    measures_specific[measure] = (
        measures_specific["total_numerator"] / measures_specific["Enrollment"]
    ) * 100

    # Drop unnecessary columns and round the percentage
    measures_specific = measures_specific.drop(
        columns=["total_numerator", "Enrollment"]
    )
    measures_specific[measure] = measures_specific[measure].round(2)

    return measures_specific
```

```python
measure_medical = calculate_specific_percentage(merged_measures, 5, "Medical")
measure_medical
```

```python
measure_medicare = calculate_specific_percentage(merged_measures, 6, "Medicare")
measure_medicare
```

```python
measure_heart = calculate_specific_percentage(merged_measures, 9, "Heart_Attack")
measure_heart
```

```python
measure_cancer = calculate_specific_percentage(merged_measures, 14, "Cancer")
measure_cancer
```

```python
measure_diabetes = calculate_specific_percentage(merged_measures, 20, "Diabetes")
measure_diabetes
```

```python
measure_kidney = calculate_specific_percentage(merged_measures, 15, "Kidney")
measure_kidney
```

```python
measure_inpatient = measures[measures["measure_id"] == 32]
measure_inpatient
```

```python
measure_inpatient = (
    measure_inpatient.groupby(["Year", "County"])
    .agg(
        total_numerator=("measure_numerator", "sum"),
        total_denominator=("measure_denominator", "sum"),
    )
    .reset_index()
)

measure_inpatient["Inpatient"] = (
    measure_inpatient["total_numerator"] / measure_inpatient["total_denominator"]
) * 12000
measure_inpatient = measure_inpatient.drop(
    columns=["total_numerator", "total_denominator"]
)
measure_inpatient["Inpatient"] = measure_inpatient["Inpatient"].round(2)
measure_inpatient
```

```python
datasets = [
    measures_age,
    measure_medical,
    measure_medicare,
    measure_heart,
    measure_cancer,
    measure_diabetes,
    measure_kidney,
    measure_inpatient,
]

cleaned_measures = datasets[0]

for df in datasets[1:]:
    cleaned_measures = cleaned_measures.merge(df, on=["Year", "County"], how="outer")

cleaned_measures
```

```python
cleaned_measures.to_csv("data/cleaned_data/measures.csv", index=False)
```
